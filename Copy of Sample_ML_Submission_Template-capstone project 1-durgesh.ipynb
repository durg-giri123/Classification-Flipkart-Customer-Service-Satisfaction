## 3. ***Data Wrangling***

Data Wrangling is the process of cleaning, transforming, and preparing raw data into a suitable format for analysis and machine learning.

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create Dataset Copy

df_copy = df.copy()
````

````python
# Convert Date Columns

df_copy['Issue_reported at'] = pd.to_datetime(
    df_copy['Issue_reported at'],
    dayfirst=True,
    errors='coerce'
)

df_copy['issue_responded'] = pd.to_datetime(
    df_copy['issue_responded'],
    dayfirst=True,
    errors='coerce'
)

df_copy['Survey_response_Date'] = pd.to_datetime(
    df_copy['Survey_response_Date'],
    errors='coerce'
)

df_copy['order_date_time'] = pd.to_datetime(
    df_copy['order_date_time'],
    dayfirst=True,
    errors='coerce'
)
````

````python
#Verify Datetime conversion
df_copy[
    [
        'Issue_reported at',
        'issue_responded',
        'Survey_response_Date',
        'order_date_time'
    ]
].head()
````

````python
# Create Response Time Feature
# Create Response Time in Minutes

df_copy['response_time_mins'] = (
    df_copy['issue_responded']
    -
    df_copy['Issue_reported at']
).dt.total_seconds() / 60
````

````python
# Verify New Feature
df_copy[
    [
        'Issue_reported at',
        'issue_responded',
        'response_time_mins'
    ]
].head()
````

````python
# Create Response Time Buckets

df_copy['response_bucket'] = pd.cut(
    df_copy['response_time_mins'],
    bins=[-1,5,15,30,60,10000],
    labels=[
        '0-5 Min',
        '5-15 Min',
        '15-30 Min',
        '30-60 Min',
        '60+ Min'
    ]
)
````

````python
# Verify bucket
df_copy['response_bucket'].value_counts()
````

````python
# Create Month-Year Feature
df_copy['Month_Year'] = (
    df_copy['Issue_reported at']
    .dt.to_period('M')
    .astype(str)
)
````

````python
# Verify
df_copy[
    [
        'Issue_reported at',
        'Month_Year'
    ]
].head()
````

### What all manipulations have you done and insights you found?

Answer Here.

Several data wrangling operations were performed to prepare the dataset for analysis. First, a copy of the original dataset was created to preserve the raw data. Date-related columns such as Issue_reported at, issue_responded, Survey_response_Date, and order_date_time were converted into datetime format to enable time-based analysis.

A new feature named response_time_mins was created by calculating the difference between issue response time and issue reporting time. This feature helped measure support efficiency and became one of the most important variables influencing customer satisfaction.

To simplify analysis, response times were grouped into meaningful response buckets. Additionally, a Month_Year feature was generated from the issue reporting date to support trend analysis over time.

These transformations improved data quality, enabled richer exploratory analysis, and created meaningful features for machine learning model development.

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

Research Question

Does response time significantly affect customer satisfaction (CSAT Score)?

Null Hypothesis (H₀)

There is no significant difference in average CSAT scores across different response time buckets.

Alternate Hypothesis (H₁)

There is a significant difference in average CSAT scores across different response time buckets.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

groups = [
    group['CSAT Score'].values
    for name, group in df_copy.groupby('response_bucket')
]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.
One-Way ANOVA Test


##### Why did you choose the specific statistical test?

Answer Here.
One-Way ANOVA was selected because the objective is to compare the mean CSAT scores across multiple response time groups.
